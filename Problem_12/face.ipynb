```python
from deepface import DeepFace
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img_path = "../Lenna.png"
img = cv.imread(img_path)
assert not img is None
```

```python
def show(img_):
    plt.figure(figsize=(10, 10), dpi=50)
    plt.imshow(img_[..., ::-1])
    plt.axis('off');

show(img)
```

```python
face_data = DeepFace.analyze(img_path, actions=["age", "gender"])
```

```python
print("Age: ", face_data["age"], "({})".format(face_data["gender"]))
```

```python
region = face_data["region"]
pt1 = (region["x"], region["y"])
pt2 = (pt1[0] + region["w"], pt1[1] + region["h"])

cv.rectangle(img, pt1, pt2, (0, 255, 0), 10)

show(img)
```
